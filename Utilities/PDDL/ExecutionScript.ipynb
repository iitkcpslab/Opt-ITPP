```python
initDir = 'problems/problems_itmp_varyRT/9/'
skip = 1
for mode in [1]:
    for robots in range(2, 4):
        for tasks in range(robots, robots+3):
            for example in range(20):
                if(robots == 2 and tasks == 4 and example == 17):
                    skip = 0
                if(skip):
                    continue
                example_id = 'problem_' + str(example).zfill(2) + '.json'
                debug_id = 'problem_' + str(example).zfill(2) + '_mode' + str(mode) + '_pddl_debug.json'
                output_id = 'problem_' + str(example).zfill(2) + '_mode' + str(mode) + '_pddl_out.json'
               
                inputLoc = os.path.join(initDir, str(robots), str(tasks), example_id)
                debugLoc = os.path.join(initDir, str(robots), str(tasks), debug_id)
                outputLoc = os.path.join(initDir, str(robots), str(tasks), output_id)
                print(inputLoc)
                print(debugLoc)
                print(outputLoc)
                print(datetime.datetime.now())
                try:
                    pddl_execute('domain.pddl', inputLoc, debugLoc, 'enhsp-dist/enhsp.jar', 0)
                except Exception as e:
                    print("Timeout ", e)
                print('Planning Ended')
                try:
                    res = pddl_output_parser(debugLoc, 'enhsp-dist/enhsp.jar')
                except Exception as e:
                    print("Failed Output Parsing ", e)
                    res = 'Failed Output Parsing\n'
                with open(outputLoc, 'w+') as fp:
                    json.dump(res, fp)
                print("Output Pasing End\n\n")
```
