```python
def get_memory():
    with open('/proc/meminfo', 'r') as mem:
        free_memory = 0
        for i in mem:
            sline = i.split()
            if str(sline[0]) in ('MemFree:', 'Buffers:', 'Cached:'):
                free_memory += int(sline[1])
    return free_memory
```

```python
import pandas as pd
import json
import os
import subprocess
```

```python
initDir = 'problems/problems_intermediate_varyZ/20/'
for mode in [1]:
    for robots in range(2, 4):
        for tasks in range(robots, robots + 2):
            if(tasks % robots):
                minZ = int(max(1, 4 + 2 * (tasks // robots)))
            else:
                minZ = int(max(1, 2 + 2 * (tasks // robots)))
            for currZ in range(minZ, minZ+5, 2):
                for example in range(10):
                    example_id = 'problem_' + str(example).zfill(2) + '.json'
                    output_id = 'problem_' + str(example).zfill(2) + '_mode' + str(mode) + '_Z' + str(currZ) + '.json'
                    inputLoc = os.path.join(initDir, str(robots), str(tasks), example_id)
                    outputLoc = os.path.join(initDir, str(robots), str(tasks), output_id)



                    execute_commands = [
                        './build/bin/taskPlanner',
                        '--config', str(inputLoc),
                        '--opt', str(mode),
                        '--numZ', str(currZ),
                        '--result', str(outputLoc),
                        '--screen', '0',
                        '--timeout', '3600'
                    ]
                    print(execute_commands)
                    try:
                        subprocess.run(execute_commands, timeout=3600)
                    except:
                        print("Failed\n")
                    print()
```
